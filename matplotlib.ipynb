```python
student ={'taha':89,'usama':56,'tayyab':78,'faseeh':86, 'ali':90}
names=list(student.keys())
marks =list(student.values())

plt.bar(names,marks)
plt.title('STUDENTS MARKS')
plt.xlabel('NAMES OF STUDENTS')
plt.ylabel('STUDENT TOTAL MARKS')

plt.show()
```

HORIZONTAL BARCHART


```python
student ={'Mohammad':58,'kashif':99,'luv':78,'shah':85, 'mujtaba':90}
names=list(student.keys())
marks =list(student.values())
plt.barh(names,marks,color='b')
plt.title('NAMES OF STUDENTS')
plt.xlabel('NAMES')
plt.ylabel('MARKS')
plt.show()
```

# **SCATTER PLOT**

```python
plt.scatter(names,marks,color='r',s=200)
plt.show()
```

# ***BOXPLOT***

```python
one =[1,2,3,2,1]
two =[3,5,7,6,2]
three= [6,3,5,6,2]

box =list([one,two,three])

plt.boxplot(box)
plt.title('boxplotiing')
plt.show()
```

```python
plt.violinplot(box)
plt.show()
```

```python
plt.violinplot(box,showmedians=True)
plt.show()
```

# ***PIE CHART***

```python
fruits =['apple','mango','banana','dates','strawbery','cherry','nuts','mirchi','aloo','tamatar']
quantity =[25,12,23,45,14,5,25,8,6,11]

plt.pie(quantity,labels= fruits)
plt.show()
```

```python
#for addding percentage
plt.pie(quantity,labels= fruits,autopct='%0.2f%%',colors =['green','yellow','blue','red','cyan','orange','black','pink','purple','grey'])
plt.show()
```
