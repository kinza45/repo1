```python


from matplotlib import pyplot as plt
import matplotlib as mpl
import numpy as np 
import random
mpl.rcParams['font.sans-serif'] = ['SimHei']   #设置简黑字体
mpl.rcParams['axes.unicode_minus'] = False # 解决‘-’bug

plt.figure()
x= np.random.rand(1000)
plt.hist(x,20)
plt.show()
```

```python

x= [1000,2000,3000]
explode=[0.01,0.03,0.01]


label=['x','y','z']

plt.figure(figsize=(20,8),dpi=80)
plt.pie(x,explode=explode,labels=label,
                            autopct='%1.1f%%',
                            colors=['y','r','c'],
                            labeldistance=1.1,
                            pctdistance=0.3)
plt.show()
```
